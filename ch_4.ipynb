## Policy Evaluation


```python
# define Environment
class Environment:
    def __init__(self, n_states=10, goal_state=9):
        self.nS = n_states
        self.goal_s = goal_state
```

```python
# Iterative Policy Evaluation
class DP_Agent:
    def __init__(self, env: Environment):
        self.env = env              # Environment
        self.gamma = 0.9            # Discount factor
        self.theta = 1e-2           # Convergence threshold
        # self.V = np.zeros(env.nS)   # State-value function
        self.V = np.random.rand(env.nS)
        # set value of goal state to 0
        self.V[env.goal_s] = 0

    def policy_evaluation(self, policy):
        while True:
            delta = 0
            for s in range(self.env.nS):
                v = self.V[s]
                new_v = 0
                for a, action_prob in enumerate(policy[s]):
                    for s_prob, next_state, reward, done in self.env.P[s][a]:
                        new_v += action_prob * s_prob * (reward + self.gamma * self.V[next_state])
                self.V[s] = new_v
                delta = max(delta, abs(v - new_v))
            if delta < self.theta:
                break
```
